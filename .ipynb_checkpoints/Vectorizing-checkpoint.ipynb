```python
import gensim
from gensim import corpora, models
import langdetect
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
```

```python
tokenizer = RegexpTokenizer(r'\w+')
```

```python
# compile sample documents into a list
#doc_set = [doc_a, doc_b, doc_c, doc_d, doc_e]
# create sample documents
input_file = "outFull.txt" #+ i
doc_set = open(input_file, 'r')

# list for tokenized documents in loop
texts = []
```

```python
# create English stop words list
lang = langdetect.detect(input_file)
lang_stop = get_stop_words('es')

# Create p_stemmer of class PorterStemmer
p_stemmer = PorterStemmer()
#print lang_stop
```

```python
# loop through document list
for i in doc_set:
    # clean and tokenize document string
    raw = i.lower()
    tokens = tokenizer.tokenize(raw)

    # remove stop words from tokens
    stopped_tokens = [i for i in tokens if not i in lang_stop]

    # stem tokens
    stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]

    # add tokens to list
    texts.append(stemmed_tokens)
    
print tokens
```

```python
#print texts
```

```python
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
#print (dictionary.token2id) 
```

```python
# generate LDA model
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=10, id2word=dictionary, passes=10)
listDicTopics = []
#dicTopic = dict()
for i in ldamodel.show_topics(len(dictionary)):
    print i
    listWords = i[1].rstrip().split('+')
    dicTopic = {i[0]:listWords}
    listDicTopics.append(dicTopic)
    
#print(corpus[0])
```

```python
#lists
word_topics = []
temp = []
dictList = []

for dicTopic in listDicTopics:
    for k,v in dicTopic.items():
        print "Topic #%s ->"%(k)
        for x in v:
            word_topics.append(x.split('*'))
            temp = [k,word_topics]
            dictList.append(temp)
            print "%s"%(x)
            
for n in dictList:
    print n
```

```python
import pandas as pd
from pandas import DataFrame

df = pd.DataFrame(dictList, columns=['Topic', 'Score'])
print df.shape
df.head()
```
